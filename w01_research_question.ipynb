## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
# ==========================================
# Section 3 - Quick Look at the Data
# ==========================================

import os
import subprocess
import pandas as pd

# Clone the FlyRank repository if it doesn't exist
REPO_DIR = "flyrank-ml-internship-starter"

if not os.path.isdir(REPO_DIR):
    subprocess.run([
        "git",
        "clone",
        "https://github.com/flyrank-bih/flyrank-ml-internship-starter.git"
    ], check=True)

# Move into the repository
os.chdir(REPO_DIR)

# Load the starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("========== DATASET OVERVIEW ==========")
print(f"Number of pages: {len(df)}")
print(f"Number of columns: {len(df.columns)}")

print("\n========== SUPPORTING NUMBERS ==========")

# 1. Average CTR
avg_ctr = df["ctr"].mean()
print(f"Average CTR: {avg_ctr:.4f}")

# 2. Correlation between Search Volume and Impressions
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between Search Volume and Impressions: {corr:.3f}")

# 3. Median Word Count by Trend Direction
print("\nMedian Word Count by Trend Direction:")
print(df.groupby("trend_direction")["word_count"].median())

print("\n========== FIRST 5 ROWS ==========")
display(df.head())
```
